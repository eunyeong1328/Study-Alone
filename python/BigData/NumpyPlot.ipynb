```python
#구를 입력하면 그 구에 해당하는 2012년부터 2018년 까지의 데이터를 그래프로 출력
#with를 사용
import matplotlib.pyplot as plt
%matplotlib inline
import csv
plt.rc('font',family="malgun gothic") #한글폰트 입력
col = []
y = []
gu = input("구를 입력해주세요: ")
with open("CCTV.csv") as inf: 
     f = inf.read().splitlines(True)
cctv = csv.reader(f)
title = next(cctv)
for data in cctv:
    data[0] = data[0].replace(' ',"")
    if data[0] == gu:
        col.append(data)
    
for row in col[0][3:10]: 
    if row == ' ':
        row = row.replace(' ',"0")
        y.append(int(row))
    else:
        row = row.replace(',',"")
        y.append(int(row))

plt.plot(title[3:10],y,"ro")
plt.show()
```

```python
#  전체구별로 출력
import matplotlib.pyplot as plt ##qt<-> inline
%matplotlib inline 
import csv
plt.rc('font',family="malgun gothic")
col= []
y =[]
cnt = 0
num =2
#f = open("../CCTV.csv")
with open("CCTV.csv") as f:
    cctv = csv.reader(f)
    title = next(cctv)
    for data in cctv:
        cnt+=1
        if(cnt == 1):
            continue
        col.append(data)
    for co in col:
        num+=1
        for row in co[3:10]:
            if(row!=0 and row!=''):
                row = int(float(row.replace(',',"")))   
            elif(row == ''):
                row = 0
            y.append(row)
        plt.plot(title[3:10], y ,label=co[0])
        y=[]
# f.close()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.show()
```

```python
#키보드로 동을 입력받고 그래프는 출력하기
import matplotlib.pyplot as plt
import csv
%matplotlib inline
plt.rc('font',family="malgun gothic") #한글폰트 입력
f = open("202203_202203_연령별인구현황_월간 (1).csv")
people = csv.reader(f)
title = next(people)
x = []
dong = input("동을 입력해주세요: ")

for data in people:
    if dong in data[0]:
        #print(data)
        for i in data[3:]:
            i= i.replace(' ','')
            i= i.replace(',','')
            x.append(int(i))
            
#plt.bar(range(101),x)
plt.bar(range(len(x)),x,label=dong)
plt.legend()
plt.grid(True)
plt.show()
f.close()
```

```python
#구 1개 받아서 구 그래프를 그리기
import matplotlib.pyplot as plt
import csv
%matplotlib qt
f= open("CCTV.csv")
cctv = csv.reader(f)
title = next(cctv)
x = []
gu = input("구를 입력해주세요 : ")
for data in cctv:
    #print(data[0])
    data[0] =  data[0].replace(' ','0')
    #print(data[0])
    if gu in data[0]:
        #print(data[3:]) #종로구 2012년부터 데이터 출력
        for i in data[3:]:
            if i == '' or i==',':
                i = i.replace('','0') #(null)데이터 0으로 교체
                i = i.replace(',','') #(,)데이터 0으로 교체
            #print(i) 
            x.append(int(i))
        #print(x) #다 처리해서 저장한 데이터
plt.bar(title[3:],x)
plt.show()
f.close()
```

```python
#구 2개 받아서 구 두개의 그래프를 그리기
import matplotlib.pyplot as plt
import csv
%matplotlib qt
plt.rc('font',family="malgun gothic") #한글폰트 입력
f= open("CCTV.csv")
cctv = csv.reader(f)
title = next(cctv)
x = []
x1 = []
gu = input("첫번째 구를 입력해주세요 : ")
gu1 = input("두번째 구를 입력해주세요 : ")
for data in cctv:
    #print(data[0])
    data[0] =  data[0].replace(' ','')
    #print(data[0])
    if gu in data[0]:
        #print(data[3:]) #종로구 2012년부터 데이터 출력
        for i in data[3:]:
            if i == '' or i==',':
                i = i.replace('','0') #(null)데이터 0으로 교체
                i = i.replace(',','') #(,)데이터 0으로 교체
            #print(i) 
            x.append(int(i))
        #print(x) #다 처리해서 저장한 데이터
    if gu1 in data[0]:
         for i in data[3:]:
            if i == '' or i== ',' :
                i = i.replace('','0')
                i = i.replace('','')
            x1.append(int(i))
plt.bar(title[3:],x,width = 0.3,label=gu)
plt.bar(title[3:],x1,width = 0.3,  align = 'edge',label=gu1)
plt.legend()
plt.show()
f.close()
```

```python
#구 2개 받아서 구 두개의 그래프를 그리기
import matplotlib.pyplot as plt
import csv
%matplotlib qt
plt.rc('font',family="malgun gothic") #한글폰트 입력
f= open("CCTV.csv")
cctv = csv.reader(f)
title = next(cctv)
x = []
x1 = []
gu = input("첫번째 구를 입력해주세요 : ")
gu1 = input("두번째 구를 입력해주세요 : ")
for data in cctv:
    #print(data[0])
    data[0] =  data[0].replace(' ','')
    #print(data[0])
    if gu in data[0]:
        #print(data[3:]) 
        for i in data[3:]:
            if i == '' or i==',':
                i = i.replace('','0') 
                i = i.replace(',','') 
            #print(i) 
            x.append(int(i)) # ~~ 이 값들중 하나만 왼쪽으로 치우치면 되기 때문에 ~~
        #print(x) 
    if gu1 in data[0]:
         for i in data[3:]:
            if i == '' or i== ',' :
                i = i.replace('','0')
                i = i.replace('','')
            x1.append(-int(i)) # ~~ 가운데를 중심으로 막대그래프 데이터가 하나로 나오는 것 ~~
plt.barh(title[3:],x,label=gu)
plt.barh(title[3:],x1,label=gu1)
plt.legend()
plt.show()
f.close()
```
