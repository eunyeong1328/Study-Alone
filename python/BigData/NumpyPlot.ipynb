```python
#구를 입력하면 그 구에 해당하는 2012년부터 2018년 까지의 데이터를 그래프로 출력
#with를 사용
import matplotlib.pyplot as plt
%matplotlib inline
import csv
plt.rc('font',family="malgun gothic") #한글폰트 입력
col = []
y = []
gu = input("구를 입력해주세요: ")
with open("CCTV.csv") as inf: 
     f = inf.read().splitlines(True)
cctv = csv.reader(f)
title = next(cctv)
for data in cctv:
    data[0] = data[0].replace(' ',"")
    if data[0] == gu:
        col.append(data)
    
for row in col[0][3:10]: 
    row = row.replace(',',"")
    y.append(int(row))

plt.plot(title[3:10],y,"ro")
plt.show()
```

```python
#구를 입력하면 그 구에 해당하는 2012년부터 2018년 까지의 데이터를 그래프로 출력
#with를 사용
import matplotlib.pyplot as plt
%matplotlib inline
import csv
plt.rc('font',family="malgun gothic") #한글폰트 입력
col = []
y = []
gu = input("구를 입력해주세요: ")
with open("CCTV.csv") as inf: 
     f = inf.read().splitlines(True)
cctv = csv.reader(f)
title = next(cctv)
for data in cctv:
    data[0] = data[0].replace(' ',"")
    if data[0] == gu:
        col.append(data)
    
for row in col[0][3:10]: 
    if row == ' ':
        row = row.replace(' ',"0")
        y.append(int(row))
    else:
        row = row.replace(',',"")
        y.append(int(row))

plt.plot(title[3:10],y,"ro")
plt.show()
```

```python
#  전체구별로 출력
import matplotlib.pyplot as plt ##qt<-> inline
%matplotlib inline 
import csv
plt.rc('font',family="malgun gothic")
col= []
y =[]
cnt = 0
num =2
#f = open("../CCTV.csv")
with open("CCTV.csv") as f:
    cctv = csv.reader(f)
    title = next(cctv)
    for data in cctv:
        cnt+=1
        if(cnt == 1):
            continue
        col.append(data)
    for co in col:
        num+=1
        for row in co[3:10]:
            if(row!=0 and row!=''):
                row = int(float(row.replace(',',"")))   
            elif(row == ''):
                row = 0
            y.append(row)
        plt.plot(title[3:10], y ,label=co[0])
        y=[]
# f.close()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.show()
```
